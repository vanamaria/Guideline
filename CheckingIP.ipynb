```python
import re, urllib, json, numpy, requests, socket
from json import load
from urllib.request import urlopen
from scapy.all import *
from itertools import chain
from prettytable import PrettyTable
from collections import Counter
```

```python
#First, validate the IP
#Regular expression for the IP
#IP should be like xxx.xxx.xx.xx or xxx.xxx.x.x or xx.x.x.x

digits = '''^(25[0-5]|2[0-4][0-9]|[0-1]?[0-9][0-9]?)\.(25[0-5]|2[0-4][0-9]|[0-1]?[0-9][0-9]?)\.(25[0-5]|2[0-4][0-9]|[0-1]?[0-9][0-9]?)\.(  25[0-5]|2[0-4][0-9]|[0-1]?[0-9][0-9]?)$'''

#Function to check the ip

def check(ip):
    if (re.search(digits, ip)):
        #IP is valid
        print("IP is valid", ip)
    else:
        print("IP is not valid", ip)

        
```

```python
def print_table(directory):
    #From the pcap file, get the IPs
    information=rdpcap(directory)
    srcIP=[]

    for inf in information:
        if IP in inf:
            srcIP.append(inf[IP].src)

        
    counter=Counter()
    
    for ip in srcIP:
        counter[ip]+=1
    
    table=PrettyTable(["Node's IP","Count"])

    for ip, count in counter.most_common():
        table.add_row([ip,count])
    print(table)    
```
